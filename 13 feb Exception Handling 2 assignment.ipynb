Q2. Write a python program to print Python Exception Hierarchy.

```python
#Answer:
def print_exception_hierarchy(exception_class, depth=0):
    print(" " * depth + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, depth + 2)
        
print_exception_hierarchy(Exception)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
#Answer:
'''The LookupError class is used as a base class for all the exceptions that occur when a specified key or index in a sequence is not found.
KeyError and IndexError are two common exceptions that are derived from the LookupError class.'''

# KeyError:

my_dict = {'a': 1, 'b': 2, 'c': 3}
try:
    value = my_dict['d']
except KeyError as e:
    print("KeyError:", e)

# IndexError:

my_list = [1, 2, 3]
try:
    value = my_list[3]
except IndexError as e:
    print("IndexError:", e)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

```python
# Answer:
'''In Python, ImportError is an exception that occurs when the import statement tries to import a module that cannot be found or loaded.
This can happen if the module name is misspelled, the module is not installed in the system, 
or if the module file is not located in the proper directory.
In Python 3.6 and later versions, a new exception called ModuleNotFoundError was introduced to handle the case when a module could not be found.
This exception is a subclass of ImportError and provides more specific information about the missing module.'''
```

Q6. List down some best practices for exception handling in python. 

```python
# Answer:

'''1. Use specific exception types: Catching specific exceptions can help you write better code that is easier to debug. 
   It is better to catch specific exceptions than to use a generic Exception type.
2. Use finally block for cleanup tasks: The finally block is executed whether or not an exception is raised, 
   so it is a good place to perform cleanup tasks such as closing files, database connections, etc.
3. Keep error messages simple: Error messages should be concise, clear, and to the point.
   They should provide enough information to help the user understand what went wrong and how to fix it.
4. Log exceptions: Logging exceptions can help you diagnose problems with your code. Use a logging framework like logging to record exceptions and other errors.
5. Use custom exceptions: Creating your own exception classes can help you handle errors in a more granular way, and provide a better experience for users.'''
```
